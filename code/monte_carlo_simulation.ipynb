```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
sims = 1_000_000

A = np.random.normal(5, 3, sims)
B = np.random.normal(20, 5, sims)
C = np.random.exponential(10, sims)
D = np.random.uniform(5, 50, sims)

X = (
    A +
    B +
    C + 
    D
)
```

```python
plt.figure(figsize=(8, 8))
plt.hist(X, density=True)
plt.axvline(10, color='r')
plt.show()

print((C > 10).sum())
```
