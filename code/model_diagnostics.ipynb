```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.linear_model import LinearRegression
```

# P-Value

```python
# Define the observed z-score
z_observed = 2.1
```

## Two Tail

```python
# Generate a range of z-values for the plot
z_values = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(z_values)

# Plot the normal distribution curve
plt.plot(z_values, pdf_values, color='black')

# Fill the area representing the p-value (two-tailed)
plt.fill_between(z_values, pdf_values, where=(z_values > z_observed), color='red', alpha=0.5, label='p-value (right tail)')
plt.fill_between(z_values, pdf_values, where=(z_values < -z_observed), color='red', alpha=0.5, label='p-value (left tail)')

# Add a vertical line for the observed z-score
plt.axvline(z_observed, color='blue', linestyle='--', label=f'z')
plt.axvline(-z_observed, color='blue', linestyle='--')

# Add labels and title
plt.title('Two-Tailed p-value for z = 2.1')
plt.xlabel('z')
plt.ylabel('Probability Density')
plt.legend()

# Adjust layout and show plot
plt.tight_layout()

plt.savefig('../content/vis/p_value_two_tail.png', dpi=300, transparent=True)
plt.show()
```

## One Tail

```python
# Generate a range of z-values for the plot
z_values = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(z_values)

# Plot the normal distribution curve
plt.plot(z_values, pdf_values, color='black')

# Fill the area representing the p-value (one-tailed, right tail)
plt.fill_between(z_values, pdf_values, where=(z_values > z_observed), color='red', alpha=0.5, label='p-value')

# Add a vertical line for the observed z-score
plt.axvline(z_observed, color='blue', linestyle='--', label=f'z')

# Add labels and title
plt.title('One-Tailed p-value for z = 2.1')
plt.xlabel('z')
plt.ylabel('Probability Density')
plt.legend()

# Adjust layout and show plot
plt.tight_layout()
plt.savefig('../content/vis/p_value_one_tail.png', dpi=300, transparent=True)
plt.show()
```

# $R^2$

## Residual Sum of Squares (SSE)

```python
# Generate some data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Fit the regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Calculate residuals
residuals = y - y_pred

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Data Points')
plt.plot(X, y_pred, color='blue', linewidth=2, label='Regression Line')

# Plot residuals
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='red', linestyle='--', linewidth=1)

plt.title('$SSR$ (Residual Sum of Squares)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.savefig('../content/vis/residual_sum_squares.png', dpi=300, transparent=True)

plt.show()
```

# Total Sum of Squares (SST)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Calculate mean of y
y_mean = np.mean(y)

# Flatten arrays for easier handling in plotting
X_flat = X.flatten()
y_flat = y.flatten()

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X_flat, y_flat, color='black', label='Data Points')
plt.axhline(y_mean, color='green', linewidth=2, label='Mean Line')

# Draw vertical lines from each data point to the mean line
for xi, yi in zip(X_flat, y_flat):
    plt.plot([xi, xi], [yi, y_mean], color='red', linestyle='--')

plt.title('$SST$ (Total Sum of Squares)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.savefig('../content/vis/total_sum_squares.png', dpi=300, transparent=True)

plt.show()
```

## Regression Sum of Squares (SSR)

```python
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Fit the regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Calculate mean of y
y_mean = np.mean(y)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Data Points')
plt.plot(X, y_pred, color='blue', linewidth=2, label='Regression Line')
plt.axhline(y_mean, color='green', linewidth=2, label='Mean Line')

# Draw vertical lines from the mean line to the regression line
for xi, yi, yi_pred in zip(X.flatten(), y.flatten(), y_pred.flatten()):
    plt.plot([xi, xi], [yi_pred, yi], color='gray', linestyle='--', alpha=0.5)

# Draw vertical lines from the regression line to the mean line
for xi, yi_pred in zip(X.flatten(), y_pred.flatten()):
    plt.plot([xi, xi], [y_mean, yi_pred], color='red', linestyle='--', alpha=0.5)

plt.title('$SSR$ (Regression Sum of Squares)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.savefig('../content/vis/regression_sum_squares.png', dpi=300, transparent=True)

plt.show()
```
