```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Unique Solution (Consistent and Independent)

```python
# Set up the figure and subplots
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# y = 2x + 1  (Line 1)
# y = -x + 4  (Line 2)

x_vals = np.linspace(-10, 10, 400)
y1 = 2 * x_vals + 1  # Line 1: y = 2x + 1
y2 = -x_vals + 4     # Line 2: y = -x + 4

ax.plot(x_vals, y1, label=r'$y = 2x + 1$', color='b')
ax.plot(x_vals, y2, label=r'$y = -x + 4$', color='r')

# Add labels, title, legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)

plt.tight_layout()

plt.savefig('../content/vis/system_linear_unique_solution.png', dpi=300)

plt.show()
```

# No Solution (Inconsistent)

```python
# Set up the figure and subplots
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# y = 2x + 1  (Line 1)
# y = 2x - 3  (Line 2)

y3 = 2 * x_vals + 1   # Line 1: y = 2x + 1
y4 = 2 * x_vals - 3   # Line 2: y = 2x - 3 (parallel)

ax.plot(x_vals, y3, label=r'$y = 2x + 1$', color='b')
ax.plot(x_vals, y4, label=r'$y = 2x - 3$', color='r')

# Add labels, title, legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)

# Adjust layout and show the plots
plt.tight_layout()

plt.savefig('../content/vis/system_linear_eq_no_solution.png', dpi=300)

plt.show()
```

# Infinitely Many Solutions (Consistent and Dependent)

```python
# Set up the figure
fig, ax = plt.subplots(figsize=(6, 6))

# --------------------------
# System with infinitely many solutions
# Equations:
# y = 2x + 3  (Line 1)
# 2y = 4x + 6 (Line 2, which is just a multiple of Line 1)

x_vals = np.linspace(-10, 10, 400)
y1 = 2 * x_vals + 3    # Line 1: y = 2x + 3
y2 = 2 * (2 * x_vals + 3) / 2  # Line 2: 2y = 4x + 6, which simplifies to y = 2x + 3

# Plot both lines (they are the same line, so they overlap)
ax.plot(x_vals, y1, label=r'$y = 2x + 3$', color='b')
ax.plot(x_vals, y2, label=r'$2y = 4x + 6$', color='r', linestyle='--')

# Add labels, title, legend
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add horizontal and vertical lines at x = 0 and y = 0
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)

# Show the plot
plt.tight_layout()

plt.savefig('../content/vis/system_linear_eq_infinite_solution.png', dpi=300)

plt.show()
```

# Multiple Equations

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.linalg import solve

# Define the equations
# Coefficients matrix (A) and constant vector (B)
A = np.array([
    [1, 1, 1],
    [2, -1, 3],
    [3, 4, -1]
])
B = np.array([6, 14, 1])

# Solve the system of equations to find the intersection point
intersection_point = solve(A, B)

# Extract coordinates of the intersection point
x_int, y_int, z_int = intersection_point

# Define the ranges for x and y
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
x, y = np.meshgrid(x, y)

# Define the equations
z1 = 6 - x - y
z2 = (14 - 2*x + y) / 3
z3 = 1 - 3*x - 4*y

# Set up the figure and 3D axes
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the planes
ax.plot_surface(x, y, z1, alpha=0.5, rstride=100, cstride=100, color='b', label='x + y + z = 6')
ax.plot_surface(x, y, z2, alpha=0.5, rstride=100, cstride=100, color='r', label='2x - y + 3z = 14')
ax.plot_surface(x, y, z3, alpha=0.5, rstride=100, cstride=100, color='g', label='3x + 4y - z = 1')

# Plot the intersection point
ax.scatter(x_int, y_int, z_int, color='red', s=100, label='Intersection Point')

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('System of 3 Linear Equations with a Unique Solution')

# Add a legend (use labels as reference)
ax.legend()

# Show the plot
plt.show()
```
