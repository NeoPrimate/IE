```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive, HBox, VBox

# Function to calculate and plot C_p, C_pk, P_p, and P_pk
def plot_cp_pp(USL, LSL, mu_process, sigma_process):
    # Calculate C_p and C_pk
    Cp = (USL - LSL) / (6 * sigma_process)
    Cpk = min((USL - mu_process) / (3 * sigma_process), (mu_process - LSL) / (3 * sigma_process))

    # P_p and P_pk assume long-term performance, use same formulae for demonstration
    Pp = Cp  # For the example, we use the same sigma as for Cp, but in practice, it's different.
    Ppk = Cpk

    # Define the range for x-axis
    x = np.linspace(LSL - 0.1, USL + 0.1, 1000)

    # Normal distribution curves
    y = (1 / (sigma_process * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu_process) / sigma_process)**2)

    # Create subplots
    fig, axs = plt.subplots(2, 2, figsize=(16, 10))

    # Plot Cp
    axs[0, 0].plot(x, y, label='Process Distribution (Cp)', color='blue')
    axs[0, 0].axvline(USL, color='red', linestyle='--', label=f'USL ({USL} mm)')
    axs[0, 0].axvline(LSL, color='green', linestyle='--', label=f'LSL ({LSL} mm)')
    axs[0, 0].axvline(mu_process, color='purple', linestyle='-', label=f'Process Mean ({mu_process} mm)')
    axs[0, 0].fill_between(x, y, where=(x > USL), color='red', alpha=0.3)
    axs[0, 0].fill_between(x, y, where=(x < LSL), color='green', alpha=0.3)
    axs[0, 0].set_title(f'Process Capability (Cp = {Cp:0.2f})')
    axs[0, 0].set_xlabel('Diameter (mm)')
    axs[0, 0].set_ylabel('Probability Density')
    axs[0, 0].legend()
    axs[0, 0].grid(True)

    # Plot Cpk
    axs[0, 1].plot(x, y, label='Process Distribution (Cpk)', color='blue')
    axs[0, 1].axvline(USL, color='red', linestyle='--', label=f'USL ({USL} mm)')
    axs[0, 1].axvline(LSL, color='green', linestyle='--', label=f'LSL ({LSL} mm)')
    axs[0, 1].axvline(mu_process, color='purple', linestyle='-', label=f'Process Mean ({mu_process} mm)')
    axs[0, 1].fill_between(x, y, where=(x > USL), color='red', alpha=0.3)
    axs[0, 1].fill_between(x, y, where=(x < LSL), color='green', alpha=0.3)
    axs[0, 1].set_title(f'Process Capability Index (Cpk = {Cpk:0.2f})')
    axs[0, 1].set_xlabel('Diameter (mm)')
    axs[0, 1].set_ylabel('Probability Density')
    axs[0, 1].legend()
    axs[0, 1].grid(True)

    # Plot Pp (for simplicity, assumed same as Cp here)
    axs[1, 0].plot(x, y, label='Process Distribution (Pp)', color='blue')
    axs[1, 0].axvline(USL, color='red', linestyle='--', label=f'USL ({USL} mm)')
    axs[1, 0].axvline(LSL, color='green', linestyle='--', label=f'LSL ({LSL} mm)')
    axs[1, 0].axvline(mu_process, color='purple', linestyle='-', label=f'Process Mean ({mu_process} mm)')
    axs[1, 0].fill_between(x, y, where=(x > USL), color='red', alpha=0.3)
    axs[1, 0].fill_between(x, y, where=(x < LSL), color='green', alpha=0.3)
    axs[1, 0].set_title(f'Process Performance (Pp = {Pp:0.2f})')
    axs[1, 0].set_xlabel('Diameter (mm)')
    axs[1, 0].set_ylabel('Probability Density')
    axs[1, 0].legend()
    axs[1, 0].grid(True)

    # Plot Ppk (for simplicity, assumed same as Cpk here)
    axs[1, 1].plot(x, y, label='Process Distribution (Ppk)', color='blue')
    axs[1, 1].axvline(USL, color='red', linestyle='--', label=f'USL ({USL} mm)')
    axs[1, 1].axvline(LSL, color='green', linestyle='--', label=f'LSL ({LSL} mm)')
    axs[1, 1].axvline(mu_process, color='purple', linestyle='-', label=f'Process Mean ({mu_process} mm)')
    axs[1, 1].fill_between(x, y, where=(x > USL), color='red', alpha=0.3)
    axs[1, 1].fill_between(x, y, where=(x < LSL), color='green', alpha=0.3)
    axs[1, 1].set_title(f'Process Performance Index (Ppk = {Ppk:0.2f})')
    axs[1, 1].set_xlabel('Diameter (mm)')
    axs[1, 1].set_ylabel('Probability Density')
    axs[1, 1].legend()
    axs[1, 1].grid(True)

    plt.tight_layout()
    plt.show()

# Define interactive widgets
USL_slider = widgets.FloatSlider(value=10.2, min=9.0, max=11.0, step=0.01, description='USL')
LSL_slider = widgets.FloatSlider(value=9.8, min=8.0, max=10.0, step=0.01, description='LSL')
mu_process_slider = widgets.FloatSlider(value=10.0, min=9.0, max=11.0, step=0.01, description='mu_process')
sigma_process_slider = widgets.FloatSlider(value=0.05, min=0.01, max=0.1, step=0.001, description='sigma_process')

# Create interactive plot
interactive_plot = interactive(plot_cp_pp,
                               USL=USL_slider,
                               LSL=LSL_slider,
                               mu_process=mu_process_slider,
                               sigma_process=sigma_process_slider)

interactive_plot
```
