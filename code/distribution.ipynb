```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t, binom, poisson, expon
```

# t- Distribution

```python
# Set degrees of freedom
df = 10  # Change this to explore different degrees of freedom

# Define the range for the x-axis
x = np.linspace(-5, 5, 1000)

# Calculate the t-distribution's probability density function (pdf)
y = t.pdf(x, df)

# Plot the t-distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='k')

# Remove axes, splines, and ticks
plt.gca().axis('off')

# Tight layout for clean visualization
plt.tight_layout()

# Save and display the plot
plt.savefig('../content/vis/t_distribution.png', dpi=300, transparent=True)
plt.show()

plt.show()
```

# Normal

```python
# Calculate the normal distribution's probability density function (pdf)
y_normal = norm.pdf(x)

# Plot the normal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y_normal, color='k')

# Remove axes, splines, and ticks
plt.gca().axis('off')

# Tight layout for clean visualization
plt.tight_layout()

plt.savefig('../content/vis/normal_distribution.png', dpi=300, transparent=True)

plt.show()
```

# Binomial

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.2  # Probability of success

# Create an array of possible outcomes
x = np.arange(0, n+1)

# Calculate the probability mass function (pmf) for each outcome
pmf = binom.pmf(x, n, p)

# Plot the binomial distribution
plt.figure(figsize=(8, 6))
plt.stem(x, pmf, basefmt=' ', linefmt='k-', markerfmt='ko')

# Remove axes and splines
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.savefig('../content/vis/binomial_distribution.png', dpi=300, transparent=True)

plt.show()
```

# Poisson

```python
# Parameters for the Poisson distribution
mu = 3  # Mean of the distribution

# Create an array of possible outcomes
x = np.arange(0, 10)

# Calculate the probability mass function (pmf) for each outcome
pmf = poisson.pmf(x, mu)

# Plot the Poisson distribution
plt.figure(figsize=(8, 6))
plt.stem(x, pmf, basefmt=' ', linefmt='k-', markerfmt='ko')

# Remove axes and splines
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.savefig('../content/vis/poisson_distribution.png', dpi=300, transparent=True)

plt.show()
```

# Exponential

```python
# Define the parameters
scale = 1  # scale parameter (1/lambda)

# Generate x values
x = np.linspace(0, 5, 1000)

# Compute the PDF values
pdf = expon.pdf(x, scale=scale)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, 'k-', lw=2, label='Exponential PDF')

plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)


plt.savefig('../content/vis/exponential_distribution.png', dpi=300, transparent=True)
# Display the plot
plt.show()
```
