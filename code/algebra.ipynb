```python
import numpy as np
import matplotlib.pyplot as plt

# Coefficients for the quadratic equation
a = 1
b = 7
c = 6

# Calculate the roots using the quadratic formula
D = b**2 - 4*a*c  # Discriminant
r1 = (-b + np.sqrt(D)) / (2 * a)  # First root
r2 = (-b - np.sqrt(D)) / (2 * a)  # Second root

# Set up the x values for the plot
x = np.linspace(-10, 2, 400)
y = a*x**2 + b*x + c  # Quadratic function

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')
plt.axhline(0, color='black', lw=0.5, ls='--')  # x-axis
plt.axvline(0, color='black', lw=0.5, ls='--')  # y-axis

# Mark the roots on the plot
plt.scatter([r1, r2], [0, 0], color='red', zorder=5)
# plt.text(r1, 0, f'  r1 = {r1:.2f}', verticalalignment='bottom', horizontalalignment='right')
# plt.text(r2, 0, f'  r2 = {r2:.2f}', verticalalignment='bottom', horizontalalignment='left')

# Annotations and titles
# plt.title('Illustration of Roots using Quadratic Formula')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 20)
plt.grid()
# plt.legend()
plt.show()
```
